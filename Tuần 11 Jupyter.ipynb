```python
def areAnagrams(path1: str, path2: str):
    with open(path1, "r") as file:
        str1 = file.read().strip().lower()
        list1 = []
        for c in str1:
            if c.isalpha():
                list1.append(c)
        sort_list1 = sorted(list1)
    with open(path2, "r") as file:
        str2 = file.read().strip().lower()
        list2 = [c for c in str2 if c.isalpha()]
        sort_list2 = sorted(list2)
    return sort_list1 == sort_list2
```

```python
def getMoney(path: str) -> list[int]:
    with open(path, "r") as file:
        numbers = file.read().split()
        x = len(numbers)
        for i in range(x):
            numbers[i] = int(numbers[i])
        return numbers
def rob(houses: list[int]) -> int:
    n = len(houses)
    if n == 0:
        return 0
    elif n == 1:
        return houses[0] 
    elif n == 2:
        return max(houses[0], houses[1])
    else:
        dp = [0] * n #tạo ra 1 list [0, 0, ....]
        dp[0] = houses[0]
        dp[1] = max(houses[0], houses[1])
        for i in range(2, n):
            dp[i] = max(dp[i - 1], dp[i - 2] + houses[i]) #dp[i]: số tiền lớn nhất có thể trộm tới căn nhà i
            """
            Lựa chọn 1: KHÔNG trộm nhà i
                 Vậy số tiền tối đa đến nhà i sẽ giống hệt đến nhà i-1 vì bỏ qua nhà i
                 -> Số tiền = dp[i-1]
            Lựa chọn 2: CÓ trộm nhà i
                 Đã trộm nhà i thì không được trộm nhà i-1 và nhà gần nhất có thể trộm trước đó là i-2
                 -> Số tiền = dp[i-2] + houses[i]
            → Luôn chọn phương án nhiều tiền hơn
            ==>lấy max của 2 lựa chọn trên
            """
        return dp[-1]                                  
rob(getMoney("houses.txt"))
```

```python
def maximumProduct(path: str) -> int:
    with open(path, "r") as file:
        numbers = list(map(int, file.read().split()[1:]))
        numbers.sort() #sắp xếp theo thứ tự tăng dần
        n = len(numbers)
        case1 = numbers[n - 1] * numbers[n - 2] * numbers[n - 3] #TH có toàn số nguyên dương hoặc nguyên âm sẽ lấy 3 số nguyên âm/ dương lớn nhất
        case2 = numbers[0] * numbers[1] * numbers[n - 1] #TH có cả số âm và dương sẽ lấy 2 số âm nhỏ nhất và 1 số dương lớn nhất
        return max(case1, case2)
```

```python
def average(path: str):
    with open(path, "r") as file:
        lines = file.read().splitlines()[1:]
        times = []
        for line in lines:
            time = float(line.split(',')[1]) #time is float
            times.append(time)
        return f"{sum(times) / len(times):.2f}"
```

Để đưa cho khách hàng các lựa chọn, rạp chiếu phim treo các poster về phim cùng với điểm đánh giá IMDB (rating) và mô tả của khách hàng về bộ phim đó.

Hãy viết hàm def findMovies (path: str) -> list[str] giúp khán giả tìm ra phim với điểm đánh giá lớn hơn 8.0 và có mô tả Không phải là "nhàm chán" ("boring").

Hàm nhận đầu vào là đường dẫn đến tệp chứa các poster phim. Hàm trả lại các tên những phim thỏa mãn điều kiện trên. Nếu không tìm thấy phim nào phù hợp, trả về thông báo "All are boring!". Kết quả in theo thứ tự xuất hiện trước sau trong file đầu vào

Tệp có cấu trúc như sau:

id, movie, description, rating

1, Thor, great 3D, 8.0

2, The Rings, Fantasy, 7.5

3, Irish, boring, 6.2

4, Ice song, Fantasy, 8.6

5, Home Alone, Interesting, 6.5

```python
def findMovies(path: str) -> list[str]:
    movies = []
    with open(path, "r") as file:
        lines = file.read().splitlines()[1:]
        movies = []
        for line in lines:
            parts = [p.strip() for p in line.split(',')] 
            if float(parts[3]) > 8.0 and parts[2].lower() != "boring":
                movies.append(parts[1])
    if not movies:
        return "All are boring!"
    return movies
```

## W11A16
Viết chương trình, nhập vào đường dẫn đến một tệp. Tệp này chứa thông tin của các hình vuông, hình tròn và hình chữ nhật. In ra chu vi của các hình trong tệp trên từng dòng, theo thứ tự xuất hiện trong tệp. Kết quả được làm tròn đến hai chữ số thập phân sau dấu phẩy.
Một tệp ví dụ như sau:

TYPE: SQUARE
SIDE: 5.3

TYPE: RECTANGLE
WIDTH: 5.0 HEIGHT: 16.0

TYPE: CIRCLE                                                                                                             RADIUS: 2.15

Một tệp có thể chứa thông tin của một hoặc nhiều hình và không theo thứ tự nhất định. Mỗi hình sẽ được định nghĩa trên hai dòng, dòng đầu tiên là loại hình (SQUARE, RECTANGLE, CIRCLE) và dòng tiếp theo là các thông số tương ứng với hình đó (SIDE, WIDTH, HEIGHT, RADIUS). Giữa thông tin của hai hình sẽ có một dòng trống.

```python
import math
def perimeter(path):
    with open(path, "r") as file:
        lines = [line for line in file.read().splitlines() if line != ""] #lines = ["TYPE: SQUARE","SIDE: 5.3","TYPE: RECTANGLE","WIDTH: 5.0 HEIGHT: 16.0","TYPE: CIRCLE", "RADIUS: 2.15"]
    n = len(lines)
    i = 0
    while i < n:
        shape = lines[i].split(":")[1].strip() 
        data = lines[i + 1]
        if shape == "SQUARE":
            side = float(data.split(':')[1].strip())
            premiter = 4 * side
        elif shape == "RECTANGLE":
            parts = data.split() #LỖI Ở ĐÂY: parts = data.split().strip() becuz split() trả về list,mà list thì KHÔNG có .strip() ⇒ sẽ lỗi AttributeError.
            width = float(parts[1])
            height = float(parts[3])
            premiter = 2 * (width + height)
        elif shape == "CIRCLE":
            radius = float(data.split(":")[1].strip())
            premiter = 2 * math.pi * radius
        print(f"{premiter:.2f}")
        i += 2
```

## W11A17
Với một dãy gồm n số nguyên với n>1, giá trị của từng phần tử tại vị trí i được tính lại bằng tích của tất cả các phần tử còn lại trong dãy số nguyên trừ chính bản thân nó - ProductofArrayExceptSelf 

Viết hàm def productExceptSelf (path: str) -> list[int]: tính và in ra màn hình dãy số nguyên sau khi được tính lại theo qui tắc trên. 

Hàm nhận đầu vào là đường dẫn path đến tệp chứa dãy số nguyên. Trả lại dãy sau khi biến đổi.

```python
import math
def productExceptSelf(path: str) -> list[int]:
    with open(path, "r") as file:
        numbers = list(map(int, file.read().split()))
    result = []
    n = len(numbers)
    for i in range(n):
        tich = math.prod(numbers[:i] + numbers[(i + 1):]) #prod: tính tích 1 list
        result.append(tich)
    return result
```

```python
#C2
def productExceptSelf(path: str) -> list[int]:
    with open(path, "r") as file:
        nums = list(map(int, file.read().split()))
    n = len(nums)
    left = [1] * n
    right = [1] * n
    result = [1] * n
    for i in range(1, n):
        left[i] = left[i-1] * nums[i-1]
    for i in range(n-2, -1, -1):
        right[i] = right[i+1] * nums[i+1]
    for i in range(n):
        result[i] = left[i] * right[i]
    return result
```

## W11A18
Bạn nhận được một tệp văn bản chứa tọa độ điểm xuất phát, tọa độ mục tiêu và vận tốc của một tên lửa (giả sử tên lửa bay thẳng với vận tốc không đổi). Hãy viết một hàm tính thời gian tên lửa cần để bay tới tiêu diệt mục tiêu của mình: def estimatedTime(path: str). Hàm nhận đầu vào là xâu kí tự chứa đường dẫn tới tệp thiết lập tên lửa và trả về số giây tên lửa hoàn thành hành trình. Đơn vị đo khoảng cách là km và đo tốc độ là km/s.

Tệp thiết lập mẫu có cấu trúc như sau:

SRC X: 0 Y: 0

DES X: 10000 Y: 10000

VELOCITY: 6 KM/S

Lấy kết quả tới 2 chữ số phần thập phân

```python
import math
def estimatedTime(path: str):
    with open(path, "r") as file:
        lines = file.read().splitlines()
    src = lines[0].split() #['SRC', 'X:', '0', 'Y:', '0']
    x_src = float(src[2])
    y_src = float(src[4])
    des = lines[1].split()
    x_des = float(des[2])
    y_des = float(des[4])
    dis = math.sqrt((x_des - x_src) ** 2 + (y_des - y_src) ** 2)
    v = float(lines[2].split()[1])
    return f"{dis / v:.2f}"
    
```

Viết chương trình nhận vào từ bàn phím một chuỗi kí tự là đường dẫn đến một tệp nhị phân. Tệp này chỉ chứa các số thực (kiểu double). In ra các số thực đó (làm tròn đến 5 chữ số thập phân sau dấu phẩy), mỗi số trên một dòng.

```python
import struct
def readBinaryNumber(path):
    with open(path, "rb") as file: #rb: đọc nhị phân
        while True:
            data = file.read(8) #read(8) = đọc 8 byte tiếp theo trong file; 8 byte = 64 bit
            if not data:
                break
            number = struct.unpack('d', data)[0] #struct giúp chuyển byte nhị phân → kiểu dữ liệu Python (VD: 8 byte → double, 4 byte → int, 1 byte → char)
                                                 #'d' = giải mã 8 byte theo chuẩn double
                                                 #struct.unpack('d', data): Chuyển 8 byte → tuple , Luôn trả về tuple, dù chỉ có 1 số
                                                 #[0] để: Vì kết quả là tuple (3.14,) lấy ra số 3.14
            print(f"{number:.5f}")
```

Nếu một điểm ảnh có ít hơn 8 điểm ảnh xung quanh, thì hãy sử dụng giá trị trung bình nhiều điểm ảnh xung quanh nhất có thể.

Viết hàm def imageSmoother (path: str) thực hiện công việc trên.

Hàm nhận đầu vào là đường dẫn đến tệp chứa file ảnh. Hàm thực hiện việc làm mịn ảnh và in ra ảnh sau khi làm mịn.

Dòng đầu tiên trong tệp chứa kích thước M của ma trận hai chiều. M dòng tiếp theo mỗi dòng chứa M pixel ứng với thông tin của ảnh.

Chú thích: Giá trị trong ma trận cho trước nằm trong khoảng [0, 255].

Chiều dài và chiều rộng của ma trận cho trước nằm trong khoảng [1, 150].

```python
def imageSmoother(path: str):
    with open(path, "r") as file:
        M = int(file.readline().strip()) #readline() là phương thức của file object, dùng để: Đọc MỘT DÒNG trong file, tính từ vị trí con trỏ hiện tại và Nó trả về chuỗi (str) Chuỗi có cả ký tự xuống dòng \n trừ dòng cuối cùng của file
        matrix_1st = []
        for _ in range (M):
            row = list(map(int, file.readline().split()))
            matrix_1st.append(row)
    matrix_2nd = [[0] * M for _ in range (M)]
    for i in range (M):
        for j in range (M):
            total = 0
            count = 0
            for dx in [-1, 0 , 1]:
                for dy in [-1, 0 , 1]:
                    x = i + dx
                    y = j + dy
                    if 0 <= x < M and 0 <= y < M: #Kiểm tra pixel có hợp lệ không: Tránh truy cập ra ngoài ma trận; Quan trọng với pixel ở biên/góc
                        total += matrix_1st[x][y]
                        count += 1
            matrix_2nd[i][j] = total // count #Làm tròn xuống
    for row in matrix_2nd:
        print(*row)
```
