## W10A3

```python
def happy_number(a):
    list1 = list(str(a))
    list2 = []
    n = len(list1)
    for i in range(n):
        list2.append(int(list1[i]))
    sum1 = 0
    for num in list2:
        sum1 += num ** 2
    a = sum1
    try:
        if happy_number(a) == 1:
            return "YES"
    except:
        return "NO"
a = int(input())
print(happy_number(a))
```

```python
try:
    a = int(input())
    if a <= 0:
        raise ValueError
    seen = set()
    while a != 1:
        if a in seen:
            print("NO")
            break
        seen.add(a)
        tong = 0
        while a > 0:
            x = a % 10
            tong += x ** 2
            a //= 10
        a = tong
    if a == 1: #or else: print("YES") becuz Trong Python, while … else có quy tắc này: Khối else CHỈ chạy khi vòng lặp kết thúc TỰ NHIÊN, tức là KHÔNG gặp lệnh break.
        print("YES")
except ValueError:
    print("Nhap sai kieu du lieu")
        
```

## W10A4

```python
try:
    a, b = map(int, input().split())
except ValueError:
    print("Sai kieu du lieu")
else: #Vì else trong try–except có ý nghĩa rất riêng: else chỉ chạy khi try KHÔNG hề xảy ra lỗi.
    print(a + b)
finally:
    print("Ket thuc chuong trinh")
```

```python
try:
    n = int(input())
    if n <= 0:
        raise ValueError
    list1 = []
    for i in range(n):
        number = int(input())
        if number <= 0:
            raise ValueError
        list1.append(number)
    if len(set(list1)) == n:
        print(sum(list1))
    else:
        print("Mang khong hop le")
except:
    print("Mang khong hop le")
```

```python
import re
t = int(input())
for i in range(t):
    s = input()
    try:
        if re.search(r'[\*\+\?]{2,}', s):
            print("False")
        else:
            re.compile(s)
            print("True")
    except re.error:
        print("False")
```
