## W5A3

```python
def isPrime(x):
    if x < 2:
        return False
    else:
        for i in range(2, int(x ** 0.5) + 1):
            if x % i == 0:
                return False
    return True
n = int(input())
print(isPrime(n))
```

## W5A4
Số hoàn hảo là số nguyên dương mà tổng của tất cả các ước số nguyên dương thật sự (proper divisors) của nó bằng chính số đó.

Ước số thật sự (Proper Divisors): Là tất cả các ước số nguyên dương của số đó, ngoại trừ chính số đó.

```python
def perfect_number(x):
    tong = 0
    for i in range (1, x):
        if x % i == 0:
            tong += i
    return (tong == x)
n = int(input())
print(perfect_number(n))
```

## W5A5

```python
def vi_tri(chuoi, k):
    for i in range (0, len(chuoi)):
        if k == chuoi[i]:
            return (i + 1)
    return -1
chuoi = list(map(int, input().split()))
k = int(input())
print(vi_tri(chuoi, k))    
```

## W5A6

```python
def factorial(n):
    tich = 1
    for i in range (1, n + 1):
        tich *= i
    return tich
n = int(input())
print(factorial(n))
```

## W5A8

```python
def hamming(a, b):
    xor_result = a ^ b
    khoang_cach = bin(xor_result).count('1')
    #bin(): chuyển sang hệ nhị phân binary string
    #.count(): đếm số lần kí tự '1' xuất hiện trong chuỗi nhị phân tạo ra
    return khoang_cach
a, b = map(int, input().split())
print(hamming(a, b))
```

## W5A9

```python
def tong_chu_so(n):
    tong = 0
    while n > 0:
        x = n % 10
        tong += x
        n = n // 10
    return tong
n = int(input())
print(tong_chu_so(n))
```

## W5A10

```python
def dang_cau(s, t):
    if len(s) != len(t):
        return False
    map_s_to_t = {}
    map_t_to_s = {}
    for char_s, char_t in zip(s, t):
        if char_s in map_s_to_t:
            if map_s_to_t[char_s] != char_t:
                return False
        elif char_t in map_t_to_s:
            return False
        else:
            map_s_to_t[char_s] = char_t
            map_t_to_s[char_t] = char_s
    return True
s, t = input().split()
print(dang_cau(s, t))
```
