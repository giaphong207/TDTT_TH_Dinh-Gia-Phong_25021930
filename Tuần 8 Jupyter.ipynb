```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def dien_tich_toan_phan(self):
        return 2 * 3.14 * ((self.r ** 2) + self.r * self.h)
    def the_tich(self):
        return 3.14 * (self.r ** 2) * self.h
r, h = map(float, input().split())
hinh_tru = Cylinder(r, h)
print(f"{hinh_tru.dien_tich_toan_phan():.2f} {hinh_tru.the_tich():.2f}")
```

## W8A2

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day 
        self.month = month
        self.year = year
    def is_leap_year(self):
        if self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0):
            return True
        return False
    def max_day(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            max_day = 31
        elif self.month in [4, 6, 9, 11]:
            max_day = 30
        else:
            if self.is_leap_year():
                max_day = 29
            else:
                max_day = 28
        return max_day
    def is_valid(self):
        if self.year < 0:
            return False
        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.max_day():
            return False
        return True
    def next_day(self):
        if not self.is_valid():
            return "INVALID"
        if self.day == 31 and self.month == 12:
            return f"01/01/{self.year + 1:04d}"
        if self.day < self.max_day():
            return f"{self.day + 1:02d}/{self.month:02d}/{self.year:04d}"
        return f"01/{self.month + 1:02d}/{self.year:04d}"
day, month, year = input().split("/")
day = int(day)
month = int(month)
year = int(year)
date = Date(day, month, year)
print(date.next_day())
```

```python
import math 
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def test(self):
        if self.x == 0 and self.y == 0:
            return "M là gốc toạ độ"
        else:
            if self.x == 0:
                return "M nằm trên trục tung"
            elif self.y == 0:
                return "M nằm trên trục hoành"
            else:
                return "M không nằm trên trục toạ độ"
    def distance(self):
        return math.sqrt(self.x ** 2 + self.y ** 2)
M = Point(*map(float, input().split()))
print(f"{M.test()} {M.distance():.2f}")         
```

```python
class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add(self):
        return self.x + self.y
    def subtract(self):
        return self.x - self.y
    def product(self):
        return self.x * self.y 
    def divide(self):
        if self.y == 0: 
            return "Lỗi: Không thể chia cho 0"
        return self.x / self.y
    def power(self):
        return self.x ** self.y
    def mod(self):
        if self.y == 0: 
            return "Lỗi: Không thể chia cho 0"
        return self.x % self.y
    def set_numbers(self, x, y):
        self.x = x
        self.y = y
x, y = map(int, input().split())
calc = Calculator(x, y)
while True:
    print("\n=======MENU=======")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia lấy dư")
    print("7. Nhập 2 số mới")
    choice = input("Chọn chức năng: ")
    match choice:
        case "1":
            print(f"\nKết quả là: {calc.add()}")
        case "2":
            print(f"\nKết quả là: {calc.subtract()}")
        case "3":
            print(f"\nKết quả là: {calc.product()}")
        case "4":
            print(f"\nKết quả là: {calc.divide()}")
        case "5":
            print(f"\nKết quả là: {calc.power()}")
        case "6":
            print(f"\nKết quả là: {calc.mod()}")
        case "7":
            a, b = map(int, input().split())
            calc.set_numbers(a, b)
            print("\nĐã cập nhật kết quả")
        case _:
            print("\nLựa chọn không hợp lệ")
    if input("Bạn có muốn thoát? ").lower() == "yes":
        print("\nKết thúc chương trình!")
        break
```

## W8A5

```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    def exists(self, name):
        for item in self.items:
            if item["name"] == name:
                return True
        return False
    def add_product(self, name, price):
        self.items.append({"name": name, "price": price})
    def remove_product(self, name):
        self.items = [item for item in self.items if item["name"] != name]
    def is_empty(self):
        return len(self.items) == 0
    def total_price(self):
        tong_tien = 0
        for item in self.items:
            tong_tien += item["price"]
        return tong_tien
    def show_items(self):
        if self.is_empty():
            print("Giỏ hàng đang rỗng")
            return
        for item in self.items:
                print (f"name: {item["name"]} - price: {item["name"]}")
    def clear_cart(self):
        self.items.clear()
cart = ShoppingCart()
while True:
    print("\n=======MENU=======")
    print("2. Thêm vào giỏ hàng sản phẩm mới")
    print("3. Xóa khỏi giỏ một sản phẩm")
    print("4. Kiểm tra xem giỏ hàng có rỗng không")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ sản phẩm trong giỏ")
    print("7. Xóa toàn bộ giỏ hàng")
    choice = input("Chọn chức năng (2 - 7): ")
    match choice:
        case "2":
            name = input("Nhập tên sản phẩm: ")
            price = float(input("Nhập giá sản phẩm: "))
            if not cart.exists(name):
                cart.add_product(name, price)
                print("\nĐã thêm sản phẩm")
            else:
                print("\nSản phẩm đã có trong giỏ hàng")
        case "3":
            name = input("Nhập tên sản phẩm cần xoá: ")
            if cart.exists(name):
                cart.remove_product(name)
                print("\nĐã xoá sản phẩm")
            else:
                print("\nSản phẩm không có trong giỏ hàng")
        case "4":
            if cart.is_empty():
                print("\nGiỏ hàng rỗng")
            else:
                print("\nGiỏ hàng không rỗng")
        case "5":
            print("\nTổng tiền:", cart.total_price())
        case "6":
            cart.show_items()
        case "7":
            cart.clear_cart()
            print("\nĐã xoá toàn bộ giỏ hàng")
        case _:
            print("\nLựa chọn không hợp lệ")
    if input("\nCó muốn thoát chương trình không? ").lower() == "yes":
        print("\nKết thúc chương trình!")
        break
```

```python
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k
    def area(self):
        return self.w * self.h * self.k ** 2
    def perimeter(self):
        return 2 * (self.w + self.h) * self.k
    def scale(self, k):
        self.w *= k
        self.h *= k
w, h, k = map(float, input().split())
hcn = Rectangle(w, h, k)
print(f"{hcn.area():.2f} {hcn.perimeter():.2f}")  
```

```python
import math
class Fraction:
    def __init__(self, num, den):
        self.num = num
        self.den = den
        self.rut_gon()
    def rut_gon(self):
        ucln = math.gcd(abs(self.num), abs(self.den))
        self.num //= ucln
        self.den //= ucln
        if self.den < 0:
            self.num = - self.num
            self.den = - self.den
    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)
    def __str__(self):
        return f"{self.num}/{self.den}"
a, b, opt, c, d = input().split()
a, b, c, d = map(int, (a, b, c, d))
phan_so_1 = Fraction(a, b)
if opt == "+":
    phan_so = phan_so_1.add(Fraction(c, d))
elif opt == "-":
    phan_so = phan_so_1.sub(Fraction(c, d))
elif opt == "*":
    phan_so = phan_so_1.mul(Fraction(c, d))
elif opt == "/":
    phan_so = phan_so_1.div(Fraction(c, d))
print(phan_so)
```

## W8A8

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, a):
        self.balance += a
        return self.balance
    def withdraw(self, b):
        if b <= self.balance:
            self.balance -= b
            return self.balance
        else:
            return "Không đủ tiền!"
owner, balance = input().split()
balance = float(balance)
account = BankAccount(owner, balance)
n = int(input())
for i in range (n):
    giao_dich, tien = input().split()
    tien = float(tien)
    if giao_dich == "DEPOSIT":
        balance = account.deposit(tien)
    elif giao_dich == "WITHDRAW":
        balance = account.withdraw(tien)
print(f"{balance:.2f}")
```

## W8A9

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}
    def add_score(self, subject, score):
        self.scores[subject] = score
    def average(self):
        return sum(self.scores.values()) / len(self.scores)
    def rank(self):
        avg = self.average()
        if avg >= 8:
            return "Excellent"
        elif avg >= 6.5:
            return "Good"
        elif avg >= 5:
            return "Average"
        else:
            return "Poor"
name = input()
student = Student(name)
n = int(input())
for i in range (n):
    subject, score = input().split()
    score = float(score)
    student.add_score(subject, score)
avg = student.average()
rank = student.rank()
print(f"{name} {avg:.2f} {rank}")
        
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library(Book):
    def __init__(self):
        self.books = []
    def add(self, book):
        self.books.append(book)
    def count_by_author(self, name):
        count = 0 
        for book in self.books:
            if book.author == name:
                count += 1
        return count
    def count_by_year(self, year):
        count = 0
        for book in self.books:
            if book.year == year:
                count += 1
        return count
n = int(input())
library = Library()
for i in range(n):
    line = input()
    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        library.add(Book(title, author, year))
    elif line.startswith("COUNTYEAR"):
        _, year = line.split()
        print(library.count_by_year(int(year)))
    elif line.startswith("COUNT"):
        _, name = line.split(" ", 1) #vì tên tác giả có thể có dấu cách
        print(library.count_by_author(name))
```

```python
import math
class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def vector_diem_moi(self, other):
        x_moi = self.x - other.x
        y_moi = self.y - other.y
        z_moi = self.z - other.z
        return Vector(x_moi, y_moi, z_moi)
class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def length(self):
        return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)
    def tich_vo_huong(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def vtpt(self, other):
        return Vector(self.y * other.z - other.y * self.z, self.z * other.x - other.z * self.x, self.x * other.y - other.x * self.y)
class Angle:
    def __init__(self, A, B, C, D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
    def calculate(self):
        AB = self.A.vector_diem_moi(self.B)
        AC = self.A.vector_diem_moi(self.C)
        n1 = AB.vtpt(AC)
        BC = self.B.vector_diem_moi(self.C)
        BD = self.B.vector_diem_moi(self.D)
        n2 = BC.vtpt(BD)
        cos_phi = abs(n1.tich_vo_huong(n2)) / (n1.length() * n2.length())
        cos_phi = max(-1, min(1, cos_phi))
        phi = math.degrees(math.acos(cos_phi))
        return phi
A = Point(*map(float, input().split()))
B = Point(*map(float, input().split()))
C = Point(*map(float, input().split()))
D = Point(*map(float, input().split()))
angle = Angle(A, B, C, D)
print(f"{angle.calculate():.2f}")
```

```python
class ComplexNumber:
    def __init__(self, realpart, impart):
        self.realpart = realpart
        self.impart = impart
    def multiply(self, other):
        real = self.realpart * other.realpart - self.impart * other.impart
        imag = self.realpart * other.impart + self.impart * other.realpart
        return ComplexNumber(real, imag)
r1, im1 = map(float, input().split())
r2, im2 = map(float, input().split())
z1 = ComplexNumber(r1, im1)
z2 = ComplexNumber(r2, im2)
result = z1.multiply(z2)
print(result.realpart, result.impart, end = " ")
```

```python
import math
class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
    def getArea(self):
        if self.e1 + self.e2 > self.e3 and self.e1 + self.e3 > self.e2 and self.e2 + self.e3 > self.e1:
            p = (self.e1 + self.e2 + self.e3) / 2
            s = math.sqrt(p * (p - self.e1) * (p - self.e2) * (p - self.e3))
            return s
        else:
            return -1
e1, e2, e3 = map(float, input().split())
tri = Triangle(e1, e2, e3)
area = tri.getArea()
if area == -1:
    print("Invalid")
else:
    print(f"{area:.2f}")
```
