Không chạy vòng lặp đệ quy vô hạn

```python
def merge(left, right):
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*merge(a, b))
```

```python
n, m = map(int, input().split())
matran = []
for  i in range (n):
    while True:
        hang = list(map(int, input().split()))
        if len(hang) != m:
            continue 
        matran.append(hang)
        break
k = int(input())
numbers = []
for i in range(n):
    for j in matran[i]:
        numbers.append(j)
def merge(left, right):
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res
def merge_sort(arr): 
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    sorted_left = merge_sort(left)
    sorted_right = merge_sort(right)
    return merge(sorted_left, sorted_right)
numbers = merge_sort(numbers)
print(numbers[k-1])
```

## W9A3

```python
def swap_to_smallest(s):
    a = list(s)
    n = len(a)
    for i in range (n):
        min_digit = a[i]
        pos = -1
        for j in range (i + 1, n):
            if a[j] < min_digit:
                min_digit = a[j]
                pos = j
            elif a[j] == min_digit and a[j] < a[i]:
                pos = j #nếu có nhiều số nhỏ giống nhau thì lấy vị cuối của số nhỏ đó trong chuỗi số
        if pos != -1 and min_digit < a[i]:
            a[i], a[pos] = a[pos], a[i] #phải dùng a[pos] mà ko đc dùng a[j] vì đã thoát vòng lặp for j
            res = ''.join(a).lstrip('0') #lstrip là xoá phía bên trái
            return res if res != "" else "0" #return: thoát ngay khỏi hàm nên chỉ swap 1 lần
    a[-1], a[-2] = a[-2], a[-1]  #nếu ko đổi được thì đổi 2 cs cuối
    return ''.join(a)
s = input().strip()
print(swap_to_smallest(s))        
```

## W9A4

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        if a[i - 1].islower():
            dp[i][0] = dp[i - 1][0]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if a[i - 1].upper() == b[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            if a[i - 1].islower():
                dp[i][j] = dp[i][j] or dp[i - 1][j]
    return "YES" if dp[n][m] else "NO"
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(can_transform(a, b))
```

```python
def kiem_tra_day_con_lien_tiep(list1, list2):
    k = len(list1)
    n = len(list2)
    if k > n:
        return "NO"
    for i in range (n - k + 1): #đã giải thích
        if list2[i] == list1[0]:
            ok = True
            for j in range (1, k):
                if list2[i + j] != list1[j]:
                    ok = False
                    break
            if ok == True:
                return "YES"
    return "NO"
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print(kiem_tra_day_con_lien_tiep(list1, list2))
```

```python
#vÒnG lẶp
def canPalaceFlowers(flowerbed, k):
    n = len(flowerbed)
    if k > n:
        return False
    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i - 1] == 0) #có ngoặc để đọc đk (có thể có hoặc ko)
            right = (i == n - 1 or flowerbed[i + 1] == 0)
            if left and right: #left và right là boolean
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
    return False
flowerbed = list(map(int, input().split()))
k = int(input())
print(canPalaceFlowers(flowerbed, k))
```

```python
#Đệ QuY
def canPlaceFlowers(flowerbed, k, i=0):
    if k == 0:
        return True
    if i >= len(flowerbed):
        return False

    if flowerbed[i] == 0:
        left = (i == 0 or flowerbed[i-1] == 0)
        right = (i == len(flowerbed)-1 or flowerbed[i+1] == 0)

        if left and right:
            flowerbed[i] = 1
            if canPlaceFlowers(flowerbed, k-1, i+2):
                return True
            flowerbed[i] = 0  # backtrack

    return canPlaceFlowers(flowerbed, k, i+1)
```

## W9A8

```python
def is_path_crossing(moves):
    x = y = 0
    path = set()
    path.add((0,0))
    for move in moves:
        if move == "R":
            x +=1
        elif move == "L":
            x -=1
        elif move == "U":
            y +=1
        elif move == "D":
            y -=1
        if (x, y) in path:
            return True
        path.add((x, y))
    return False
moves = input()
print(is_path_crossing(moves))
```

## W9A9

```python
def Median(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d: 
        c, d = d, c
    if a > c: 
        a, c = c, a
    if b > d: 
        b, d = d, b
    if b > c: 
        b, c = c, b
    if e < b:
        return b
    elif e > c:
        return c
    else:
        return e
nums = list(map(int, input().split()))  
print(Median(nums))
```
