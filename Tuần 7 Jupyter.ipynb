## W7A1

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right :
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid + 1
    return -1 
arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, target))
```

## W7A3

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range (0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
arr = list(map(int,input().split()))
print(bubble_sort(arr))
    
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
arr = list(map(int,input().split()))
print(selection_sort(arr))
```

## W7A4

```python
def so_lan_nhieu_nhat(arr):
    m = 0 
    for i in arr:
        n = arr.count(i)
        m = max(m, n)
    return m     
arr = list(map(int, input().split()))
m = so_lan_nhieu_nhat(arr)
list_1 = []
for i in arr:
    if arr.count(i) == m and i not in list_1:
        list_1.append(i)
q = len(arr)
for j in list_1:
    p = arr.index(j)
    q = min(q, p)
print(f"{arr[q]} xuat hien nhieu nhat, som nhat, {m} lan")
```

## W7A13

```python
def longest_harmony_sequence(arr):
    dict_1 = {}
    for i in arr:
        if i in dict_1:
            dict_1[i] += 1
        else:
            dict_1[i] = 1
    m = 0
    for i in dict_1:
        if i + 1 in dict_1:
            n = dict_1[i] + dict_1[i+1]
            m = max(m, n)
    return m
arr = list(map(int, input().split()))
print(longest_harmony_sequence(arr))
```
