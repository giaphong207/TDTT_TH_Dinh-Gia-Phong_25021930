## W7A1

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right :
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid + 1
    return -1 
arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, target))
```

## W7A3

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range (0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
arr = list(map(int,input().split()))
print(bubble_sort(arr))
    
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
arr = list(map(int,input().split()))
print(selection_sort(arr))
```

```python
def do_dai_lis(danh_sach):
    dp = []
    for so in danh_sach:
        trai, phai = 0, len(dp)
        while trai < phai:
            giua = (trai + phai) // 2
            if dp[giua] < so:
                trai = giua + 1
            else:
                phai = giua
        if trai == len(dp):
            dp.append(so)
        else:
            dp[trai] = so
    return len(dp)
danh_sach = list(map(int, input().split()))
print(do_dai_lis(danh_sach))
```

## W7A8

```python
def tim_trai(a, x):
    l, r = 0, len(a)
    while l < r:
        m = (l + r) // 2
        if a[m][0] < x: l = m + 1
        else: r = m
    return l
def tim_phai(a, x):
    l, r = 0, len(a)
    while l < r:
        m = (l + r) // 2
        if a[m][0] <= x: l = m + 1
        else: r = m
    return l
def khoang_ngan_nhat(interval, queries):
    qs = [(queries[i], i) for i in range(len(queries))]
    qs.sort()
    kq = [-1] * len(queries)
    for l, r in interval:
        do_dai = r - l + 1
        i = tim_trai(qs, l)
        j = tim_phai(qs, r)
        for k in range(i, j):
            idx = qs[k][1]
            if kq[idx] == -1 or do_dai < kq[idx]:
                kq[idx] = do_dai
    return kq
interval = eval(input())
queries = eval(input())
print(khoang_ngan_nhat(interval, queries))
```

## W7A9


```python
def dem_nho_hon(nums):
    ds = []
    kq = [0]*len(nums)
    for i in range(len(nums)-1, -1, -1):
        x = nums[i]
        l, r = 0, len(ds)
        while l < r:
            m = (l+r)//2
            if ds[m] < x:
                l = m+1
            else:
                r = m
        kq[i] = l
        ds.insert(l, x)
    return kq
nums = list(map(int, input().split()))
print(dem_nho_hon(nums))
```

## W7A10

```python
def tim_lon_hon(ds, x):
    l, r, p = 0, len(ds)-1, -1
    while l <= r:
        m = (l+r)//2
        if ds[m] > x:
            p = m
            r = m-1
        else:
            l = m+1
    return p
def max_mod(nums, m):
    pre = 0
    best = 0
    ds = []
    for x in nums:
        pre = (pre + x) % m
        best = max(best, pre)
        i = tim_lon_hon(ds, pre)
        if i != -1:
            best = max(best, (pre - ds[i] + m) % m)
        ds.append(pre)
        j = len(ds)-1
        while j > 0 and ds[j] < ds[j-1]:
            ds[j], ds[j-1] = ds[j-1], ds[j]
            j -= 1
    return best
nums = list(map(int, input().split()))
m = int(input())
print(max_mod(nums, m))
```

## W7A13

```python
def longest_harmony_sequence(arr):
    dict_1 = {}
    for i in arr:
        if i in dict_1:
            dict_1[i] += 1
        else:
            dict_1[i] = 1
    m = 0
    for i in dict_1:
        if i + 1 in dict_1:
            n = dict_1[i] + dict_1[i+1]
            m = max(m, n)
    return m
arr = list(map(int, input().split()))
print(longest_harmony_sequence(arr))
```
