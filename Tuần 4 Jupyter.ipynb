```python
def palindrome(chuoi):
    return chuoi == chuoi[::-1]
n = int(input("Nhập vào 1 số nguyên dương n: "))
buoc = 0
while True:
    n_str = str(n)
    if palindrome(n_str):
        break
    chuoi_dao_nguoc = n_str[::-1]
    n += int(chuoi_dao_nguoc)
    buoc += 1
print(f"Số bước: {buoc}")
print(f"Giá trị palindrome: {n}")
    
```

## W4A10 

```python
def Collatz(a):
    buoc = 1
    while a != 1:
        if a % 2 == 0:
            a = a / 2
        else:
            a = 3 * a + 1
        buoc += 1
    return buoc            
n = int(input("Nhập vào 1 số nguyên dương n: "))
x_dai_nhat = 1
do_dai_lon_nhat = 1
for x in range(1, n + 1):
    do_dai_hien_tai = Collatz(x)
    if do_dai_hien_tai > do_dai_lon_nhat:
        do_dai_lon_nhat = do_dai_hien_tai
        x_dai_nhat = x
print(x_dai_nhat)
print(do_dai_lon_nhat)
```

## W4A14

```python
#UCLN
def UCLN(x,y):
    UCLN = 0
    if x <= y:
        for i in range(1,x+1):
            if x % i == 0 and y % i == 0:
                UCLN = max(UCLN,i)
        return UCLN
    if x > y:
        for i in range(1,y+1):
            if x % i == 0 and y % i == 0:
                UCLN = max(UCLN,i)
        return UCLN
m, n = map(int, input("Nhập vào 2 số nguyên dương: ").split())
ket_qua = UCLN(m,n)
print(f"Ước chung lớn nhất của m = {m} và n = {n} là: {ket_qua} ")
```

```python
n = input("Nhập vào 1 số nguyên n: ")
ket_qua = ""
so_hien_tai = int(n)
while so_hien_tai > 0:
    phan_duoi = so_hien_tai % 1000
    so_hien_tai = so_hien_tai // 1000
    if so_hien_tai > 0:
        nhom_so = f"{phan_duoi:03}" #03 để thêm 0 vào kiểu 5 thành 005 1 nhóm và 3 là giới hạn 3 nhóm 1 số
        ket_qua ="." + nhom_so + ket_qua
    else: #Khi n còn ít hơn 3 chữ số
        ket_qua = str(phan_duoi) + ket_qua
print(f"{n} thành {ket_qua}")
```
