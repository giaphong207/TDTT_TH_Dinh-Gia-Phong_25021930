## W6A3

```python
numbers = list(map(int, input().split()))
k = int(input())
if k < 0:
    print("Lỗi: K là số nguyên dương!")
do_dai = len(numbers)
if k < do_dai:
    new_list = numbers[k:] + numbers[:k]
else:
    x = k // do_dai
    y = k - x * do_dai
    # 2 dòng <=> y = k % do_dai
    new_list = numbers[y:] + numbers[:y]
print(tuple(new_list))
```

## W6A4

```python
from collections import defaultdict
keys_values = input()
pairs = keys_values.split()
result_dict = defaultdict(list)
for pair in pairs:
    key, value = pair.split(':', 1) #Dùng .split(':', 1) để đảm bảo chỉ tách ở dấu : đầu tiên, hữu ích nếu value cũng chứa dấu :
    result_dict[key].append(value)
final_dict = dict(result_dict)
print(final_dict)
```

## W6A5

```python
from collections import defaultdict
numbers = list(map(int, input().split()))
result_dict = defaultdict(int)
for i in numbers:
    if i == 0:
        result_dict['zeros'] += 1
    elif i > 0:
        result_dict['positives'] += 1
    else:
        result_dict['negatives'] += 1
final_dict = dict(result_dict)
print(final_dict)
```

## W6A9

```python
from collections import defaultdict
from ast import literal_eval
dict_1 = literal_eval(input())
dict_2 = literal_eval(input())
result_dict = defaultdict(int)
for key, value in dict_1.items():
    result_dict[key] += value
for key, value in dict_2.items():
    result_dict[key] += value
sorted_keys = sorted(result_dict.keys())
final_dict = {}
for key in sorted_keys:
    final_dict[key] = result_dict[key]
print(final_dict)
```

## W6A17

```python
n = int(input())
ma_tran = []
for i in range (n):
    while True:
        hang = list(map(int, input().split()))
        if len(hang) != n:
            continue
        ma_tran.append(hang)
        break
duong_cheo_chinh = [] #Vị trí i = j (hàng = cột)
duong_cheo_phu = [] #Vị trí i + j = n - 1
for i in range(n):
    phan_tu_chinh = ma_tran[i][i]
    duong_cheo_chinh.append(phan_tu_chinh)
    j = n - 1 - i
    phan_tu_phu = ma_tran[i][j]
    duong_cheo_phu.append(phan_tu_phu)
print(duong_cheo_chinh)
print(duong_cheo_phu)
    
```
